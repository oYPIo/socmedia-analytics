```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

2. Детально ознайомитись із можливостями бібліотек matplotlib і pywt.
3. Побудувати вейвлет-скейлограму з базовою вейвлет-функцією.

```python
t = np.linspace(0, 1, 1000, endpoint=False)
signal = np.sin(25 * 2 * np.pi * t) * np.exp(-10 * (t - 0.5)**2)

scales = np.arange(1, 128)
coeffs, freqs = pywt.cwt(signal, scales, 'cmor')

plt.figure(figsize=(10, 6))
plt.imshow(np.abs(coeffs), aspect='auto', extent=[0, 1, 1, 128], cmap='jet')
plt.colorbar(label="Амплітуда")
plt.ylabel("Масштаби")
plt.xlabel("Час")
plt.title("Вейвлет-скейлограмма (CWT)")
plt.show()
```

4. Дослідити різницю вейвлет-скейлограм при різних базових вейвлет-функціях.

```python
for wavelet in ["morl", "cmor", "mexh"]:
    coeffs, _ = pywt.cwt(signal, scales, wavelet)
    plt.figure(figsize=(10, 4))
    plt.imshow(np.abs(coeffs), aspect="auto", extent=[0, 1, 1, 128], cmap="jet")
    plt.colorbar(label="Амплітуда")
    plt.title(f"Скейлограма з вейвлетом {wavelet}")
    plt.show()
```
